# segmenation analysis

```python
datasegment = df[['Age','Weight','Height','Sleep_Duration','Heart_Rate','Blood_Oxygen_Level','Calories_Intake', 'Water_Intake','Body_Fat_Percentage','Muscle_Mass', 'Health_Score']].copy()
```

```python
datasegment.head()
```

```python
datasegment = datasegment.fillna(0.)
```

```python
import missingno as msno
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(datasegment)
```

```python
pca = PCA(random_state = 42)
pca.fit(data_scaled)
```

```python
data_pca = pca.transform(data_scaled)
df_pca = pd.DataFrame(data_pca, columns = [f'PC{x +1}' for x in range(len(datasegment.columns))])
pca_summary = pd.DataFrame({'Standard Deviation':df_pca.std(),
                           'Proportion of Variance': pca.explained_variance_ratio_,
                           'Cumultive Proportion': pca.explained_variance_ratio_.cumsum()})
pca_summary
```

```python
factor_loadings = pd.DataFrame(pca.components_, columns = datasegment.columns, index = df_pca.columns).T
```

# division of the segments and do analyis on the data

```python
factor_loadings.style.background_gradient(cmap = 'Blues')
```

```python

# Calculate centroid
index_names = factor_loadings.index

# Plot data points
sns.scatterplot(data=df_pca, x = 'PC1', y = 'PC2', color = 'blue')

# Plot arrows from centroid to provided coordinates with index names
for i in range(len(factor_loadings['PC1'])):
    plt.arrow(0, 0, factor_loadings['PC1'][i] * 3.5, factor_loadings['PC2'][i] * 3.5, head_width=0.05, head_length=0.1, fc='red', ec='red')
    plt.text((factor_loadings['PC1'][i] * 3.5), factor_loadings['PC2'][i]* 3.5, index_names[i], fontsize=12, ha='center', color = 'red', va = 'center_baseline')

# Set labels and legend
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('A Principal components analysis of the list of members of the data')
plt.grid(True)
plt.savefig("pca_2w.png", format = 'png')
```

```python
data_km28 = [KMeans(n_clusters = k, random_state = 42).fit(data_scaled) for k in range(2, 9)]
wcss = [model.inertia_ for model in data_km28]
fig = plt.figure(figsize = (10,6))
sns.barplot(x = list(range(2, 9)), y = wcss, color = 'blue')
plt.xlabel("Segments")
plt.ylabel('Sum of Within Cluster Distances')
plt.title("Scree plot for the customers ")
plt.savefig("scree_plot.png")
plt.show()
```

# get the profilling segments

```python
kmeans = data_km28[2]
data_clust = pd.DataFrame(data_scaled, columns = datasegment.columns)
data_clust['cluster'] = kmeans.labels_
data_pca = df_pca.copy()
data_pca['cluster'] = kmeans.labels_
data_profile = datasegment.copy()
data_profile['cluster'] = kmeans.labels_
columns_segment = data_clust.columns
data_pivot = data_profile[columns_segment].groupby('cluster').mean().T
data_pivot_mean = data_pivot.mean(axis = 1).reset_index()
data_pivot_mean.columns = ['Variable', 'Value']
```

```python

plt.figure(figsize = (12, 9))
for i in range(4):
    plt.subplot(2, 2, i+1)
    sns.barplot(data=data_pivot, x = i, y = data_pivot.index)
    sns.scatterplot(data=data_pivot_mean, x = 'Value', y = 'Variable', color = 'red')
    for index, row in data_pivot_mean.iterrows():
        plt.hlines(y=row['Variable'], xmin=0, xmax=row['Value'], colors='red', alpha = 0.2)
    plt.ylabel("")
    plt.xlabel("")
    plt.title(f"Cluster {i}: {data_profile['cluster'].value_counts()[i]} ({data_profile['cluster'].value_counts()[i]*100/len(data_profile):.0f}%)")
plt.suptitle("Segment profile plot for the four-segment solution for the people health  data set", fontsize = 15)
plt.tight_layout()
plt.savefig("segment_profile.png")
plt.show()
```

Cluster 0 (24%): This group shows the highest average calorie intake, significantly higher than other clusters. They likely focus on high-energy diets but show average values in other health metrics.

Cluster 1 (28%): The largest cluster, with a calorie intake that is slightly lower than Cluster 0. This group maintains a balanced approach to health, with moderate levels across most health metrics.

Cluster 2 (24%): Similar in size to Cluster 0, but with lower calorie intake compared to Cluster 1. This cluster might represent individuals who are more cautious about their diet, reflected in slightly higher muscle mass.

Cluster 3 (23%): The smallest cluster, characterized by the lowest average calorie intake among all segments. They likely prioritize health-conscious behaviors, with balanced but lower metrics in other health areas.

```python
import pandas as pd

data = pd.read_csv('personal_health_data.csv')
df = data.copy()  # Creating a copy of the DataFrame is generally a good practice

# Calculate BMI and add it as a column (BMI = Weight (kg) / (Height (m)^2))
df['BMI'] = df['Weight'] / (df['Height'] / 100) ** 2

# Convert 'Stress_Level' to numeric, setting errors='coerce' to handle non-numeric values
df['Stress_Level'] = pd.to_numeric(df['Stress_Level'], errors='coerce')

# Handle NaN values by filling with 0 or other strategy
df['Stress_Level'].fillna(0, inplace=True)

# Health Risk Factor Example: combining BMI and stress level (simplified)
df['Health_Risk_Factor'] = df['BMI'] + df['Stress_Level']

# Example Engagement Level: Simplified
df['Engagement_Level'] = df['Sleep_Duration'] - df['Stress_Level']

# Calculate churn rate: Assuming 'Anomaly_Flag' = 1 means high risk of churn
# Check if 'Anomaly_Flag' exists in the DataFrame
if 'Anomaly_Flag' in df.columns:
    churn_rate = df['Anomaly_Flag'].mean()  # % of users with Anomaly_Flag = 1
else:
    print("Warning: 'Anomaly_Flag' column not found. Churn rate cannot be calculated.")
    churn_rate = 0  # Set churn rate to 0 or handle as needed

df['Revenue'] = df['Health_Score'] * 10  # Example revenue calculation
ARPU = df['Revenue'].mean()

# Calculate CLTV (Customer Lifetime Value)
# Simplified: Assume Gross Margin = 70% (can vary by business)
gross_margin = 0.7
CLTV = ARPU * gross_margin * (1 / churn_rate if churn_rate != 0 else 0)

# Print key metrics
print(f"Churn Rate: {churn_rate:.2%}")
print(f"ARPU: ${ARPU:.2f}")
print(f"CLTV: ${CLTV:.2f}")

# Print the DataFrame to verify the calculations
print(df.head())
```

 Calculation: Converts weight and height into BMI.
Stress Level Conversion: Converts Stress_Level to numeric and handles non-numeric values by converting them to NaN and then replacing NaN with 0.
Health Risk Factor: Combines BMI and Stress Level.
Engagement Level: Subtracts Stress Level from Sleep DBMIuration.
Churn Rate: Calculates the percentage of users flagged as anomalies.
ARPU and CLTV: Estimates revenue and customer lifetime value based on Health Score.
